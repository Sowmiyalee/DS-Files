```python
from scipy import stats
```

```python
stats.norm.cdf(x=70,loc=340,scale=30)  #loc is mean, scale is standard deviation, we can pass in the orde without loc and scale as well
```

```python
stats.norm.cdf(70,60,10)  #finds the probability to the left side of the curve
```

```python
1-stats.norm.cdf(70,60,10) #finds the probability to the right side of the curve
```

```python
stats.norm.ppf(0.8413447460685429,60,10)  #finds the x value while passing the probality
```

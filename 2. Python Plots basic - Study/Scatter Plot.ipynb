# **Compare Plots**

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)
plt.xlabel('Age of the car')
plt.ylabel('Speed of the car')
plt.title('COMPARISON BETWEEN CAR MODEL AND SPEED OF THE CAR')
```

**We can change the colours of data point**

```python

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color='#2c14ff')    #Colours are take from google - hue colours

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y,color = '#00AF1A')
plt.xlabel('Age of the car')
plt.ylabel('Speed of the car')
plt.title('COMPARISON BETWEEN CAR MODEL AND SPEED OF THE CAR')
```

```python
import seaborn as sns
```

```python
df=pd.read_csv('/content/Automobile_data.csv',nrows= 100)
# nrows will import only required rows
```

```python
sns.scatterplot(data=df, x='city-mpg', y ='highway-mpg')
plt.figure(figsize=(20,10))
```

```python
sns.scatterplot(data=df, x='city-mpg', y ='highway-mpg',hue='fuel-type')

plt.rcParams['figure.figsize'] = (20,10)
plt.rcParams['figure.dpi'] = 100
```

*   Ones








```python
a = np.arange(10,21)
a
```

```python
b = np.arange(100,201)
b
```

```python
c = np.linspace(1,10,20)
c
```

```python
## to slice we give rows first, then column, Viz G [Rows, Cols]

G[0,0]
```

```python
G[0,1:]
```

```python
G[:,1:]
```

```python
G[1:,:]
```

```python
G[1:,1:]
```

```python
G[1:,2]
```

```python
G[:,0:1]
```

```python
G[:,0:2]
```

```python
G[[0,0],[1,2]]
```

```python
G[[0,1],[1,2]]
```

# **Stacking**

```python
a = np.array([1,2,3])
b = np.array ([4,5,6])
```

```python
c = np.vstack([a,b])
c
```

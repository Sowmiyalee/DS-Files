# **Filtering in Pandas**

```python
df['sex'].value_counts()
```

```python
df['sex'] == 'Male'
```

```python
## filtering only male
df[df['sex'] == 'Male']
```

```python
## Salary is greater than 150000
df[df['salary'] > 150000]
```

```python
## Find Salary is greater than 150000 and Female 
## (For AND use &, for OR use | )
```

```python
df[(df['sex'] == 'Female') & (df['salary'] > 150000)]
```

```python
## What is the average salary of females?
```

```python
df[df['sex'] == 'Female']['salary'] . mean()]
```

```python
## We can assign this code to get value to a variable and call it whenever we want
```

```python
Avg_sal_female = df[df['sex'] == 'Female']['salary'] . mean()
```

```python
print("The average salary of feamale is : ", Avg_sal_female )
```

```python
## What is the average salary of rank prof and discipline is A?
```

```python
df[(df['rank'] == 'Prof') & (df['discipline'] == 'A')]['salary'] . mean()
```

# **Matplotlib Histogram**

```python
import pandas as pd 
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
plt.hist(df['salary'],bins=10)
plt.title ('salary data')
```

```python
salary_mean = df['salary'].mean()
salary_median = df['salary'].median()

plt.hist(df['salary'],bins=10)
plt.axvline(salary_mean,color='y')
plt.axvline(salary_median,color='r')

plt.xlabel('Salary')
plt.ylabel('Frequency')
```

```python
df['phd'].hist(bins=5,grid=False,alpha=0.5,histtype='bar',ec='black')
```
